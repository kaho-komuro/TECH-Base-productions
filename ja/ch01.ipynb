## 00. パタトクカシーー
2つの文字列「パトカー」と「タクシー」の文字を先頭から交互に連結し、文字列「パタトクカシーー」を得よ。

```python
linea="パトカー"
lineb="タクシー"
line=""
for i in range(len(linea)):
  line+=linea[i]+lineb[i]
print(line)
```

## 02. 文字列の逆順
文字列"stressed"の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ。

```python
line="stressed"
new_line=""
for i in range(len(line)):
  new_line+=line[len(line)-i-1]
print(new_line)
```

## 04. 元素記号
"Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."という文を単語に分解し、1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字、それ以外の単語は先頭の2文字を取り出し、取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ。

```python
sentence="Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
words=sentence.split()
numberlist=[1, 5, 6, 7, 8, 9, 15, 16, 19]
dict={}
for i in range(len(words)):
  if i+1 in numberlist:
    dict[words[i]]=words[i][0]
  else:
    dict[words[i]]=words[i][0:2]
print(dict)
```

## 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ。この関数を用い、"I am an NLPer"という文から文字tri-gram、単語bi-gramを得よ。

```python
def n_gram(n,sentence):
  if type(sentence) is str:
    se=sentence.split()
  if n==1:
    return se
  else:
    ngram=[]
    for i in range(len(se)-n+1):
      ngram.append(se[i:i+n])
    return ngram
```

```python
sentence="I am an NLPer"
print("bi-gram:",n_gram(2,sentence))
print("tri-gram:",n_gram(3,sentence))
```

## 06. 集合
"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を、それぞれ, $X$ と $Y$ として求め、$X$ と $Y$ の和集合（$X \cup Y$）、積集合（$X \cap Y$）、差集合（$X \setminus Y$）を求めよ。さらに、'se'というbi-gramがXおよびYに含まれるかどうかを調べよ。

```python
def char_bigram(sentence):
    bigram = []
    for i in range(len(sentence)-1):
        bigram.append(sentence[i:i+2])
    return bigram

x = "paraparaparadise"
y = "paragraph"

# リストとしてbi-gramを取得
list_X = char_bigram(x)
list_Y = char_bigram(y)

# リストを集合(set)に変換することで、重複が取り除かれる
X = set(list_X)
Y = set(list_Y)

print("X:", X)
print("Y:", Y)

# setを使えば、集合演算は記号一つで計算できます
union = X | Y          # 和集合 (X.union(Y) でも可)
intersection = X & Y   # 積集合 (X.intersection(Y) でも可)
difference = X - Y     # 差集合 (X.difference(Y) でも可)

print("和集合 (X ∪ Y):", union)
print("積集合 (X ∩ Y):", intersection)
print("差集合 (X - Y):", difference)

# 'se' というbi-gramがX, Yに含まれるかを調べる
# 集合の中に特定の要素があるかどうかは `in` を使います
print("'se' は X に含まれるか？:", 'se' in X)
print("'se' は Y に含まれるか？:", 'se' in Y)
```

```python
union={}
intersection={}
difference={}

for i in range(len(X)):
  if X[i] is None in union:
    union[X[i]]=X[i]
  if X[i] is None in intersection:
    intersection[X[i]]=X[i]
  if X[i] is None in difference:
    difference[X[i]]=X[i]

for i in range(len(Y)):
  if Y[i] is None in union:
    union[Y[i]]=Y[i]
  if Y[i] in intersection:
    intersection[Y[i]]=1
  if Y[i] is None in difference:
    difference[Y[i]]=Y[i]
  if Y[i] in difference:
    del difference[Y[i]]

print(X)
print(Y)
print("和集合:",union)
print("積集合:",intersection)
print("差集合:",difference)
```

## 08. 暗号文
与えられた文字列の各文字を、以下の仕様で変換する関数cipherを実装せよ。

* 英小文字ならば (219 - 文字コード) のASCIIコードに対応する文字に置換
* その他の文字はそのまま出力

この関数を用い、英語のメッセージを暗号化・復号化せよ。

```python
def cipher(sentence):
  N=len(sentence)
  st=""
  for i in range(N):
    if sentence[i].islower()==True:
      st+=str(ord(sentence[i]))
    else:
      st += sentence[i]
  return st

cipher("I am an NLPer")
```

## 09. Typoglycemia
スペースで区切られた単語列に対して、各単語の先頭と末尾の文字は残し、それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ。ただし、長さが４以下の単語は並び替えないこととする。適当な英語の文（例えば"I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."）を与え、その実行結果を確認せよ。

```python
def sentence_swap(sentence):
  se=sentence.split()
  N=len(se)
  if N<=4:
    return sentence
  else:
    new_sentence=""
    new_sentence+=se[0]
    new_sentence+=" "
    for i in range(1,N-2):
      new_sentence+=se[N-i-1]
      new_sentence+=" "
    new_sentence+=se[N-1]
  return new_sentence

sentence_swap("I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind .")
```
